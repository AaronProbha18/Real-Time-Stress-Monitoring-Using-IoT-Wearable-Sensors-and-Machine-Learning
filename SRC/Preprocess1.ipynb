```python
import statistics

mean = {
    "Pulse":   statistics.mean(df["Pulse"]),
    "Temp":   statistics.mean(df["Temp"]),
    "Ambient Noise":   statistics.mean(df["Ambient Noise"])
}

stdev = {
    "Pulse":   statistics.stdev(df["Pulse"]),
    "Temp":   statistics.stdev(df["Temp"]),
    "Ambient Noise":   statistics.stdev(df["Ambient Noise"])
}

print(mean, "\n", stdev)
```

```python
zPulse = (df["Pulse"] - mean["Pulse"])/stdev["Pulse"]
zTemp = (df["Temp"] - mean["Temp"])/stdev["Temp"]
zAmbNoise = (df["Ambient Noise"] - mean["Ambient Noise"])/stdev["Ambient Noise"]


print(zPulse,"\n", zTemp, "\n", zAmbNoise)
```

```python
df["ZPulse"]=zPulse
df["ZTemp"]=zTemp
df["ZAmbNoise"]=zAmbNoise

df.to_csv("SaafData.csv")
```
